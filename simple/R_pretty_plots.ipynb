### Creating nice looking plots in R

The R kernel by default uses fonts from the system, and some of these may be pretty low resolution. You get much nicer results when using so-called vector fonts. Vector graphics do not contain pixels, but are description of what an image should look like (e.g. "a line going from (0,0) to (1,1), with a thickness of 0.1"). This 'description' can be rendered at any resolution, without looking pixelated. The fonts from the Cairo R-package are vector fonts. This notebook shows how they can be used to create nice looking image lables.

First, let's see what is the problem with the default font used by the R kernel

```r
options(repr.plot.width = 10, repr.plot.height = 8, repr.plot.res=96)
library('ggplot2')

ggplot(iris) + geom_point(aes(Sepal.Width, Petal.Width)) + cowplot::theme_cowplot(font_size=25)
```

As we can see, the axes labels are pixelated. Using a vector font gives a much nicer result. We select the Cairo backend by setting the `repr.plot.backends` option. Then, we select the font family we want to use (in this example: FreeSans), and finally, we create the plot:

```r
options(repr.plot.width = 10, repr.plot.height = 8, repr.plot.res=96, repr.plot.backends=c('Cairo'))
library('ggplot2')
library('Cairo')
CairoFonts(
regular="FreeSans:style=Medium",
bold="FreeSans:style=Bold",
italic="FreeSans:style=Oblique",
bolditalic="FreeSans:style=BoldOblique"
)
ggplot(iris) + geom_point(aes(Sepal.Width, Petal.Width)) + cowplot::theme_cowplot(font_size=25)
```

You can query all available fonts using the following code block. Any font that has 'truetype' in the directory name, is a vector-based font and will thus look nice.

```r
install.packages("systemfonts", lib="~/JHL_installations/R/")
library('systemfonts')
options(repr.matrix.max.rows=1000)
system_fonts()
```

Just one more example, this time rendering at a higher resolution

```r
options(repr.plot.width = 10, repr.plot.height = 8, repr.plot.res=150, repr.plot.backends=c('Cairo'))
library('ggplot2')
library('Cairo')
CairoFonts(
regular="Liberation Sans Narrow:style=Medium",
bold="Liberation Sans Narrow:style=Bold",
italic="Liberation Sans Narrow:style=Oblique",
bolditalic="Liberation Sans Narrow:style=BoldOblique"
)
ggplot(iris) + geom_point(aes(Sepal.Width, Petal.Width)) + cowplot::theme_cowplot(font_size=25)
```
